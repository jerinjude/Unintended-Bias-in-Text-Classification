```python
text1=[]
for i in text_rem:
    i=str(i)
    text=re.sub(r'\[.*\]',' ',i)
    text=text.lower()
    text=re.sub(r'https:\S*','',text) #http
    text=re.sub(r'[0-9]',' ',i)
    text=re.sub(r'\n',' ',text)
    text=re.sub(r'\\',' ',text)
    text=re.sub(r'\/',' ',text)
    text=re.sub(r'\-',' ',text)
    text=re.sub(r'[\!\"\#\$\%\&\\\'\(\)\*\+\,\-\.\/\:\;\<\=\>\?\@\[\]\^\_\`\{\|\}\~]','',text)
    text=re.sub(r'\s{2:}',' ',text)
    text=re.sub(r'\s.*\\\S+\s',' ',text)
    text=re.sub(r'\b[a-z]\b','',text)
    text=text.lower()
    text=text.encode('ascii','ignore')
    text=text.decode()
    text=re.sub(r'\s+',' ',text)
    text=re.sub('\b\w\b',' ',text)
    words=text.split()
    if len(words)>3:
        text1.append(text)
```

```python
def preprocess_text(text):
    words=[contractions.fix(word) for word in text.split()]
    text=(' ').join(words)
    text=text.lower()
    stop_words = set(stopwords.words('english'))
    words=[word for word in text.split() if word not in stop_words]
    text=str((' ').join(words))
    token=spcy(text)
    text=" ".join([token.lemma_ for token in token])
    return text
```

```python
df1['clean_text']=df1['text'].apply(preprocess_text)
```

```python
df1['length']=df1['text'].apply(lambda x: len(x))
df1['no_of_words']=df1['text'].apply(lambda x: len(x.split()))
df1['avg_word_length']=df1['text'].apply(lambda x: np.mean([len(x) for x in x.split()]))
df1['caps']=df1['text'].apply(lambda x: len(re.findall(r'[A-Z]',x)))
df1['excl']=df1['text'].apply(lambda x: len(re.findall(r'\!',x)))
df1['sentence_count'] = df1['text'].apply(lambda x: len(str(x).split(".")))
df1['avg_sentence_length'] = df1['text'].apply(lambda x: np.mean([len(x) for x in x.split('.')]))
```

```python
analyzer = SentimentIntensityAnalyzer()
df1['sentiment']=df1['text'].apply(lambda x: analyzer.polarity_scores(x).get('compound'))
```

```python
def get_subjectivity(text):
    try:
        textblob = TextBlob(unicode(text, 'utf-8'))
        subj = textblob.sentiment.subjectivity
    except:
        subj = 0.0
    return subj
```

```python
df1['subjectivity']=df1['text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
pos_dic = {
    'noun' : ['NN','NNS','NNP','NNPS'],
    'pron' : ['PRP','PRP$','WP','WP$'],
    'verb' : ['VB','VBD','VBG','VBN','VBP','VBZ'],
    'adj' :  ['JJ','JJR','JJS'],
    'adv' : ['RB','RBR','RBS','WRB']
}

def pos_check(x, flag):
    x = re.sub("[^a-zA-Z]", " ", x)
    x = " ".join(x.split())
    cnt = 0
    try:
        wiki = TextBlob(x)
        for tup in wiki.tags:
            ppo = list(tup)[1]
            if ppo in pos_dic[flag]:
                cnt += 1
    except:
        pass
    return cnt
df1['noun_count'] = df1['text'].apply(lambda x: pos_check(x, 'noun'))
df1['verb_count'] = df1['text'].apply(lambda x: pos_check(x, 'verb'))
df1['adj_count'] = df1['text'].apply(lambda x: pos_check(x, 'adj'))
df1['adv_count'] = df1['text'].apply(lambda x: pos_check(x, 'adv'))
df1['pron_count'] = df1['text'].apply(lambda x: pos_check(x, 'pron'))
```
